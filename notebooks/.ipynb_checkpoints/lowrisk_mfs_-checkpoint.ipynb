# for a mutualfund with low risk shd have treynor ration between (min) & (max) 

```python
data['Treynor'].describe()
```

```python
low_risk_trenor = data[(data.Treynor.values > (data['Treynor'].min())) & (data.Treynor.values < (data['Treynor'].max()))]
```

```python
low_risk_trenor
```

```python
data['Sharpe Ratio'].describe()
```

### the ideal value of medium risk mutual fund shd have sharp rattio in between min& max


```python
mfs_sharpe = low_risk_trenor[(low_risk_trenor['Sharpe Ratio'] > (data['Sharpe Ratio'].min())) & (low_risk_trenor['Sharpe Ratio'] < (data['Sharpe Ratio'].max()))]
mfs_sharpe
```

```python
data['Standard Deviation(%)'].describe()
```

```python
std_consistency =  mfs_sharpe[(mfs_sharpe['Standard Deviation(%)'] >(0)) & (mfs_sharpe['Standard Deviation(%)'] < (5.25))]
std_consistency.sort_values(by='cagr',ascending=False)
```

```python
data['R square'].describe()
```

```python
r2_funds = std_consistency[(std_consistency['R square']>(data['R square'].mean()))&((std_consistency['R square']<(data['R square'].max())))]
r2_funds
```

```python
data['Alpha'].describe()
```

```python
alpha_mfs = r2_funds[(r2_funds['Alpha']>(data['Alpha'].min()))&(r2_funds['Alpha']<(data['Alpha'].max()))]
alpha_mfs
```

```python
data['Beta'].describe()
```

```python
beta_mfs = alpha_mfs[(alpha_mfs['Beta']>(0))&(alpha_mfs['Beta']<(6.99))]
beta_mfs.sort_values(by='cagr',ascending=False)
```
