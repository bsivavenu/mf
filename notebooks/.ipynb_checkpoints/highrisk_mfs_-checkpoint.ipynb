# for a mutualfund with high risk shd have treyorration between (mean) and  (max) that is 3 to 9

```python
data['Treynor'].describe()
```

```python
high_risk_trenor = data[(data.Treynor.values > (data['Treynor'].mean())) & (data.Treynor.values < (data['Treynor'].max()))]
```

```python
high_risk_trenor
```

```python
data['Sharpe Ratio'].describe()
```

### the ideal value of medium risk mutual fund shd have sharp rattio in between mean& max


```python
mfs_sharpe = high_risk_trenor[(high_risk_trenor['Sharpe Ratio'] > (data['Sharpe Ratio'].mean())) & (high_risk_trenor['Sharpe Ratio'] < (data['Sharpe Ratio'].max()))]
mfs_sharpe
```

```python
data['Standard Deviation(%)'].describe()
```

```python
std_consistency =  mfs_sharpe[(mfs_sharpe['Standard Deviation(%)'] >(0)) & (mfs_sharpe['Standard Deviation(%)'] < (19.68))]
std_consistency.sort_values(by='cagr',ascending=False)
```

```python
data['R square'].describe()
```

```python
r2_funds = std_consistency[(std_consistency['R square']>(data['R square'].min()))&((std_consistency['R square']<(data['R square'].max())))]
r2_funds
```

```python
data['Alpha'].describe()
```

```python
data[data['Alpha']>16]
```

```python
data['Alpha'].plot()
```

```python
alpha_mfs = r2_funds[(r2_funds['Alpha']>(data['Alpha'].min()))&(r2_funds['Alpha']<(data['Alpha'].max()))]
alpha_mfs
```

```python
data['Beta'].describe()
```

```python
beta_mfs = alpha_mfs[(alpha_mfs['Beta']>(0))&(alpha_mfs['Beta']<(6.99))]
beta_mfs.sort_values(by='cagr',ascending=False)
```
