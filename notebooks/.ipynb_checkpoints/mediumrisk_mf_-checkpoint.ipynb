## other factors to take into consideration
1. Expense Ratio - for large caps it is high and vice versa
2. number of holdings of a mf
3. longterm of mf - a mf fund with 3-5 is preferred
4. turnover ratio - it is % change of holdings of a mf - lower % change is better
