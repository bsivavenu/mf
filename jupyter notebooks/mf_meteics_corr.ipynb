if cust wants to take high risk and but mf shd be have consistency that is less std deviation

```python
data.head()
```

```python
plt.figure(figsize=(20,5))
sns.pointplot(data['mf'].values,data['Standard Deviation(%)'].values)
plt.xticks(rotation=90)
plt.show()
```

```python
high_risk_low_std = data[(data['Sharpe Ratio'] > data['Sharpe Ratio'].mean())&(data['Standard Deviation(%)']>data['Standard Deviation(%)'].mean())].sort_values(by='cagr',ascending=False,inplace=False)
```

```python
high_risk_low_std.head(10)
```

```python
# sns.pointplot(,.values)
plt.figure(figsize=(20,5))
sns.pointplot(high_risk_low_std['mf'].values,high_risk_low_std['Standard Deviation(%)'].values)
# sns.pointplot(high_risk_low_std['mf'].values,high_risk_low_std['Sharpe Ratio'].values)
sns.pointplot(high_risk_low_std['mf'].values,high_risk_low_std['cagr'].values)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
